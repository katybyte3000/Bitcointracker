# Aufschlüsselung nach Datum und Title

```python
from selenium.webdriver.common.by import By
import pandas as pd

termin = []
titles = []

for i, post in enumerate(posts):
    print(f"Scraping post {i + 1} of {len(posts)}")

    # statt running zählt es die Durchläufe


for post in posts:

    try:
        datum = post.find_element(By.CSS_SELECTOR, ".humble").text
        
    except:
        datum = ""
    termin.append(datum)

    try:
        title = post.find_element(By.CSS_SELECTOR, ".news__item-title").text
    except:
        title = ""
    titles.append(title)

# Erstelle den DataFrame
posts_df = pd.DataFrame({
    'Datum': termin,
    'Titel': titles
})

print(posts_df)
```

```python
from selenium.webdriver.common.by import By
import pandas as pd

termin = []
titles = []

for i, post in enumerate(posts):
    print(f"Scraping post {i + 1} of {len(posts)}")

    try:
        # Nur das Datum extrahieren (alles vor dem " - ")
        date_text = post.find_element(By.CSS_SELECTOR, ".humble").text
        termin.append(date_text.split(" - ")[0])
    except:
        termin.append("")

    try:
        title = post.find_element(By.CSS_SELECTOR, ".news__item-title").text
    except:
        title = ""
    titles.append(title)

# DataFrame erstellen
posts_df = pd.DataFrame({
    'Datum': termin,
    'Titel': titles
})

print(posts_df)
```
